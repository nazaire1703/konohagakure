```python
import yfinance as yf
from yahoofinancials import YahooFinancials
from yahooquery import Ticker
import pandas as pd

from bs4 import BeautifulSoup
import requests

# https://data.library.virginia.edu/getting-started-with-web-scraping-in-python/
from selenium import webdriver
import time
# https://pypi.org/project/webdriver-manager/
from webdriver_manager.chrome import ChromeDriverManager
from webdriver_manager.core.utils import ChromeType
```

```python
def rendering(url):
        # ChromeDriverManager(chrome_type=ChromeType.CHROMIUM).install()
        driver = webdriver.Chrome(ChromeDriverManager().install()) # run ChromeDriver
        driver.get(url)                                          # load the web page from the URL
        time.sleep(3)                                            # wait for the web page to load
        render = driver.page_source                              # get the page source HTML
        driver.quit()                                            # quit ChromeDriver
        return render                                            # return the page source HTML
```

```python
url = "https://investor.vanguard.com/investment-products/etfs/profile/vt"
# url = "https://www.schwab.wallst.com/schwab/Prospect/research/etfs/schwabETF/index.asp?type=holdings&symbol=VT" #portfolio-composition

wunderground_page = rendering(url)
```

```python
# html = requests.get(url)
soup = BeautifulSoup(wunderground_page, "lxml")
table = soup.find("holding-details-results").find("table")

headers = []
for i in table.find('thead').find_all('th'):
    title = i.text.strip()
    headers.append(title)
headers = headers[1:]
print(headers)

tickers = [i.text.strip() for i in table.find('tbody').find_all("th")]
print(tickers)

df = pd.DataFrame(columns=headers)

for row in table.find('tbody').find_all("tr"):
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    if len(row_data):
        df.loc[len(df),:] = row_data

df.insert(0,'Ticker', tickers)
df
```

```python
etf_list = {'SCHD':0.25, 'VOO':0.25, 'VGT':0.10, 'QQQ':0.10, 'ARKK':0.05, 'O':0.05, 'WPC':0.05, 'BTC-USD':0.05, 'ETH-USD':0.05}
df_total = pd.DataFrame()

for name, percent in zip(etf_list.keys(), etf_list.values()):
    ticker = Ticker(name)
    if type(ticker.fund_profile[name]) == dict:
        if ticker.fund_profile[name]['legalType'] == 'Exchange Traded Fund':
            df = pd.DataFrame(ticker.fund_holding_info[name]['holdings']).reset_index()
    else:
        df = pd.DataFrame({'symbol':[name], 'holdingName':[''], 'holdingPercent':[1]}).reset_index()
    df.insert(0, 'NAME', name)
    df['%'] = df['holdingPercent'] * percent

    df_total = pd.concat([df_total, df], axis=0)
    
df_agg = df_total.groupby('symbol').agg({'%':'sum'})
df_agg = df_agg.merge(df_total[['symbol', 'holdingName']], on='symbol', how='inner')
df_agg = df_agg.drop_duplicates(ignore_index=True)
df_agg = df_agg.sort_values('%', ascending=False)
df_agg.insert(2,'cum%', df_agg['%'].cumsum())
df_agg.style.format(formatter="{:.1%}", subset=['%', 'cum%'])
```
